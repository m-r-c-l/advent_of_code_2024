# Data

```python
puzzle = open('../data/day_1.txt')
puzzle_input = puzzle.read().split('\n')

list_1 = [int(line.split()[0]) for line in puzzle_input]

list_2 = [int(line.split()[1]) for line in puzzle_input]
```

# Puzzles

## Puzzle 1

```python
list_1.sort()
list_2.sort()
distances = []
```

```python
for i in range(len(list_1)):
    distances.append(abs(list_1[i]-list_2[i]))
```

```python
sum(distances)
```

## Puzzle 2

```python
occurrence = []

for i in list_1:
    occurrence.append(list_2.count(i))
```

```python
occurrence
```

```python
similarity = []

for index, value in enumerate(list_1):
    similarity.append(value * occurrence[index])
```

```python
sum(similarity)
```
